### Now Let's Transpose the data frame

```python
india_df = india_df.transpose()
india_df.head()
```

```python
india_df.rename(columns=india_df.iloc[0],inplace=True)
```

```python
# Drop the Series name too 
india_df.drop(['Series Code','Series Name'],axis=0,inplace=True)
```

```python
india_df.isna().count()
```

### Now let's check all the columns and we will choose only the ones necessary for our analysis


```python
india_df.columns
```

```python
india_df = india_df[['Population, total','Population density (people per sq. km of land area)','Life expectancy at birth, total (years)',
                            'Fertility rate, total (births per woman)','Adolescent fertility rate (births per 1,000 women ages 15-19)',
                            'GDP (current US$)','Foreign direct investment, net inflows (BoP, current US$)','Military expenditure (% of GDP)',
                            'Number of infant deaths']]
```

```python
india_df.head()
#let's change the dtype of population to millions
india_df['Population, total'] = india_df['Population, total'].apply(lambda x:x/1000000)
```

```python
plt.figure(figsize=(18,10))
plt.bar(india_df.index,india_df['Population, total'],color='white')
plt.xlabel('Years')
plt.ylabel('Population In Millions')
```

```python
plt.figure(figsize=(18,10))
plt.plot(india_df.index,india_df['Population density (people per sq. km of land area)'],linestyle='dotted')
plt.xlabel('Years')
plt.ylabel('Population Per SQ. KM of Land Area')
```

```python

sns.relplot(x=india_df.index,y=india_df['Fertility rate, total (births per woman)'],kind='line',data=india_df)
```
